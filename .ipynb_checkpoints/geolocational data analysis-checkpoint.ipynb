### Data cleaning

The dataset has 61 features of that not all are relavant, so a new dataframe is derived from the original one which consists only the features which I felt have a relavance(Based on proper justification).
The new dataframe is called df_clean, all the visualisation and all are applied on this dataframe. Now this df_clean has some important 11 features.

```python
df_clean = df[['cook','eating_out','employment','ethnic_food','exercise','fruit_day','income','on_off_campus','pay_meal_out','sports','veggies_day']]
```

There are many null values in the dataframe so all of them are dropped as they don't help us in any case while visualisation, rather create an unnecessary trouble.

```python
df_clean.dropna(inplace = True, axis = 0)
```

### Data Visualisation

Visualising the distribution of data with the help of a seaborn boxplot
